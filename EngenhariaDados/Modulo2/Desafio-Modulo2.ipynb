Atividade 1: classe MapaPaisContinente

```python
class MapaPaisContinente():
    def __init__(self):
      self.mapeamento_continente = {}

    def mapear(self, df: pd.DataFrame, col_pais: str, col_continente: str):
      for pais, continente in zip(df[col_pais], df[col_continente]):
          self.mapeamento_continente[pais] = continente
    
    def retorna_continente(self, pais:str):
      try:
        return self.mapeamento_continente[pais]
      except KeyError:
        print("Não existe esse país em nossa base")
        return None
```

```python
mapa = MapaPaisContinente()
mapa.mapear(users, 'Country or area', 'Region')
print(mapa.mapeamento_continente)
```

```python
print(mapa.retorna_continente('Brazil'))
```

```python
print(mapa.retorna_continente('Brasil'))
```
